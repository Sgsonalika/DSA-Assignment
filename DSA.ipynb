1 DIscuss string slicing and provide examples.

Ans

String slicing is a technique in programming that allows you to extract a specific portion or "slice" of a string. Imagine a string as a sequence of characters, each with a position or index, starting from 0 for the first character.

Basic Syntax

The basic syntax for slicing is:

string[start:stop:step]

i. start: The index where the slice starts (inclusive).

ii. stop: The index where the slice ends (exclusive).

iii. step: (Optional) The interval between characters in the slice.

1 Simple Slice:

text = "Hello, World!"

print(text[0:5:1]) # Output: Hello

2 Omitting Start or Stop:

text = "Hello, World!"

print(text[:5]) # Output: Hello

print(text[7:]) # Output: World!

text[:5] starts from the beginning and goes up to index 5.

text[7:] starts from index 7 and goes to the end of the string.

3 Using Step:

text = "Hello, World!"

print(text[::2]) # Output: Hlo ol!

text[::2] means "take every second character" starting from the beginning.

4 Negative Indexing:

text = "Hello, World!"

print(text[-6:-1]) # Output: World

Negative indices count from the end of the string. So text[-6:-1] starts at the 6th character from the end and goes up to (but not including) the last character.

5 Reversing a String:

text = "Hello, World!"

print(text[::-1]) # Output: !dlroW ,olleH

The slice text[::-1] reverses the string by starting from the end and moving backward with a step of -1.

Key Points

If start is omitted, slicing starts from the beginning.

If stop is omitted, slicing continues to the end.

The step is optional, and if omitted, the default is 1 (every character).

This makes string slicing a powerful tool for extracting specific parts of a string or manipulating it in various ways.

2 Explain the key features of lists in Python

Ans

Lists in Python are a versatile and powerful data structure that allows you to store and manipulate a collection of items. Here are the key features of lists:

i. Ordered Collection:

Description: Lists maintain the order of elements. The first item added to the list will be at the first position, the second item at the second position, and so on.

Example:

fruits = ["apple", "banana", "cherry"]

print(fruits[0]) # Output: apple

ii. Mutable:

Description: Lists are mutable, meaning you can change their content after they are created. You can add, remove, or modify elements.

Example

fruits[1] = "blueberry" # Change "banana" to "blueberry"

print(fruits) # Output: ['apple', 'blueberry', 'cherry']

iii. Heterogeneous Elements:

Description: A list can contain elements of different data types, including numbers, strings, and even other lists.

Example:

mixed_list = [1, "Hello", 3.14, [5, 6]]

print(mixed_list) # Output: [1, 'Hello', 3.14, [5, 6]]

iv. Dynamic Size:

Description: Lists in Python can grow or shrink in size. You can add or remove elements without worrying about declaring the size of the list in advance.

Example

numbers = [1, 2, 3]

numbers.append(4) # Add an element

numbers.remove(2) # Remove an element

print(numbers) # Output: [1, 3, 4]

v. Slicing and Indexing:

Description: You can access individual elements or a range of elements in a list using indexing and slicing.

Example

letters = ['a', 'b', 'c', 'd', 'e']

print(letters[1:4]) # Output: ['b', 'c', 'd']

vi. Built-in Functions and Methods:

Description: Python provides many built-in functions and methods to manipulate lists, such as len(), append(), pop(), sort(), and more.

Example

fruits = ["apple", "cherry", "banana"]

fruits.sort() # Sort the list

print(fruits) # Output: ['apple', 'banana', 'cherry']

vii. List Comprehensions:

Description: Python allows you to create new lists by applying an expression to each element of an existing list using a concise syntax known as list comprehension.

Example

squares = [x**2 for x in range(5)]

print(squares) # Output: [0, 1, 4, 9, 16]

viii. Supports Nesting:

Description: Lists can contain other lists as elements, allowing for the creation of complex data structures like matrices or lists of lists.

Example

matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

print(matrix[1][2]) # Output: 6

ix. Iterability:

Description: You can easily iterate over the elements of a list using loops.

Example:

for fruit in fruits:

print(fruit) #Output:
             
             apple
             
             banana
             
             cherry

3 Describe how to access, modify, and delete elements in a list with examples

Ans

To access, modify, and delete elements in a list in Python with examples.

i. Accessing Elements in a List

You can access elements in a list using their index. Python uses zero-based indexing, so the first element has an index of 0.

#Accessing a Single Element:

fruits = ["apple", "banana", "cherry"]

print(fruits[0]) # Output: apple

print(fruits[2]) # Output: cherry

#Accessing Elements from the End (Negative Indexing):

print(fruits[-1]) # Output: cherry (last element)

print(fruits[-2]) # Output: banana (second-to-last element)

#Accessing a Range of Elements (Slicing):

print(fruits[1:3]) # Output: ['banana', 'cherry']

ii. Modifying Elements in a List

Lists are mutable, meaning you can change the value of an element by assigning a new value to a specific index.

#Modifying a Single Element:

fruits[1] = "blueberry" # Change "banana" to "blueberry"

print(fruits) # Output: ['apple', 'blueberry', 'cherry']

#Modifying Multiple Elements (Using Slicing):

fruits[1:3] = ["kiwi", "mango"] # Replace "blueberry" and "cherry" with "kiwi" and "mango"

print(fruits) # Output: ['apple', 'kiwi', 'mango']

iii. Deleting Elements from a List

You can remove elements from a list using several methods, such as del, remove(), and pop().

#Deleting an Element by Index (Using del):

del fruits[1] # Remove the element at index 1

print(fruits) # Output: ['apple', 'mango']

#Removing an Element by Value (Using remove()):

fruits.remove("apple") # Remove the first occurrence of "apple"

print(fruits) # Output: ['mango']

#Removing and Returning the Last Element (Using pop()):

last_fruit = fruits.pop() # Remove and return the last element

print(last_fruit) # Output: mango

print(fruits) # Output: []

#Removing and Returning an Element by Index (Using pop(index)):\

fruits = ["apple", "banana", "cherry"]

second_fruit = fruits.pop(1) # Remove and return the element at index 1

print(second_fruit) # Output: banana

print(fruits) # Output: ['apple', 'cherry']

4 Compare and contrast tuples and lists with examples.

Ans

Tuples and lists are both data structures in Python that can store a collection of items. However, they have some key differences and similarities.

i. Mutability

Lists: Mutable, meaning you can change, add, or remove elements after the list is created.

Tuples: Immutable, meaning once a tuple is created, you cannot change, add, or remove elements.

Example:

#List

fruits_list = ["apple", "banana", "cherry"]

fruits_list[1] = "blueberry" # You can modify elements

print(fruits_list) # Output: ['apple', 'blueberry', 'cherry']

#Tuple

fruits_tuple = ("apple", "banana", "cherry")

#fruits_tuple[1] = "blueberry" # This will raise an error

ii. Syntax

Lists: Defined using square brackets [].

Tuples: Defined using parentheses ().

Example:

fruits_list = ["apple", "banana", "cherry"]

fruits_tuple = ("apple", "banana", "cherry")

iii. Use Cases

Lists: Preferred when you need a collection that may change over time, such as adding or removing items.

Tuples: Used when you need a collection that should not change, such as storing a fixed set of values like coordinates, or when you want to ensure the data remains constant.

Example

#List (Dynamic collection)

shopping_list = ["milk", "eggs", "bread"]

shopping_list.append("butter") # Adding an item to the list

print(shopping_list) # Output: ['milk', 'eggs', 'bread', 'butter']

#Tuple (Fixed collection)

point = (10, 20)

#point[0] = 15 # This would raise an error, as tuples are immutable

iv. Methods

Lists: Have more built-in methods like append(), remove(), pop(), sort(), etc.

Tuples: Have fewer methods, mainly related to counting occurrences and finding the index of elements (count(), index()).

Example:

#List methods

numbers_list = [1, 2, 3]

numbers_list.append(4) # Adding an element

numbers_list.remove(2) # Removing an element

print(numbers_list) # Output: [1, 3, 4]

#Tuple methods

numbers_tuple = (1, 2, 3, 2)

print(numbers_tuple.count(2)) # Output: 2 (counts the occurrences of 2)

print(numbers_tuple.index(3)) # Output: 2 (finds the index of 3)

v.Nesting

Lists: Can be nested inside other lists, and you can also nest tuples inside lists.

Tuples: Can be nested inside other tuples, and you can also nest lists inside tuples.

Example:

#Nested lists

nested_list = [[1, 2], [3, 4]]

#Nested tuples

nested_tuple = ((1, 2), (3, 4))

#Mixed nesting

mixed = ([1, 2], (3, 4))

Summary:

Lists are mutable, meaning they can be modified after creation, making them suitable for collections that need to change. They are defined using square brackets [] and have a wide range of methods for manipulation.

Tuples are immutable, meaning they cannot be changed after creation, making them ideal for fixed collections of items. They are defined using parentheses () and have fewer methods than lists.

5 Describe the key features of sets and provide examples of their use.

Ans

A set in Python is an unordered collection of unique elements. Sets are commonly used when you need to store a collection of items but want to ensure that there are no duplicates. Here are the key features of sets along with examples of their use:

i Unordered Collection

Description: Sets do not maintain any order of elements. The items in a set are stored in a way that makes them efficient for membership tests (checking whether an item is in the set).

Example

fruits = {"apple", "banana", "cherry"}

print(fruits) # Output might be: {'cherry', 'banana', 'apple'}

ii. Unique Elements

Description: Sets automatically eliminate duplicate values. If you add a duplicate element to a set, it will only appear once.

Example

numbers = {1, 2, 2, 3, 4, 4, 5}

print(numbers) # Output: {1, 2, 3, 4, 5}

iii. Mutable

Description: Sets are mutable, meaning you can add or remove elements after the set has been created.

Example

fruits = {"apple", "banana"}

fruits.add("cherry") # Adding an element

print(fruits) # Output: {'banana', 'cherry', 'apple'}

iv. No Indexing or Slicing

Description: Since sets are unordered, you cannot access elements by index or slice them like you can with lists or tuples.

Example

fruits = {"apple", "banana", "cherry"}

#print(fruits[0]) # This will raise an error because sets do not support indexing

v. Set Operations

Sets support mathematical operations like union, intersection, difference, and symmetric difference, which makes them very powerful for certain types of problems.

Union (| or union()): Combines all elements from both sets.

Intersection (& or intersection()): Returns only elements present in both sets.

Difference (- or difference()): Returns elements present in the first set but not in the second.

Symmetric Difference (^ or symmetric_difference()): Returns elements present in either of the sets but not in both.

Examples:

set1 = {1, 2, 3}

set2 = {3, 4, 5}

union_set = set1 | set2 # {1, 2, 3, 4, 5}

intersection_set = set1 & set2 # {3}

difference_set = set1 - set2 # {1, 2}

symmetric_diff_set = set1 ^ set2 # {1, 2, 4, 5}

print("Union:", union_set)

print("Intersection:", intersection_set)

print("Difference:", difference_set)

print("Symmetric Difference:", symmetric_diff_set)

vi. Mutable Set Methods

Sets provide several methods for modifying the set in place.

#Adding Elements (add()): Adds a single element to the set.

fruits = {"apple", "banana"}

fruits.add("orange")

print(fruits) # Output: {'apple', 'orange', 'banana'}

#Removing Elements (remove(), discard()): Removes an element from the set. If the element does not exist, remove() raises an error, while discard() does nothing.

fruits.remove("banana") # Removes "banana" from the set

fruits.discard("pear") # Does nothing, as "pear" is not in the set

print(fruits) # Output: {'apple', 'orange'}

#Clearing a Set (clear()): Removes all elements from the set.

fruits.clear()

print(fruits) # Output: set()

vii. Immutability with Frozenset

Description: Python provides a special type of set called frozenset, which is immutable. Once created, you cannot modify a frozenset (similar to how tuples relate to lists).

Example:

frozen_set = frozenset([1, 2, 3, 4])

#frozen_set.add(5) # This will raise an error because frozenset is immutable

print(frozen_set) # Output: frozenset({1, 2, 3, 4})

viii. Membership Testing

Description: Sets are optimized for checking membership, meaning they are very efficient when you need to check if an element is in a set.

Example

fruits = {"apple", "banana", "cherry"}

print("apple" in fruits) # Output: True

print("pear" in fruits) # Output: False

Ans

Dictionaries in Python are mutable data structures that store key-value pairs. Here’s how you can add, modify, and delete items in a dictionary:

i. Adding Items

You can add new key-value pairs to a dictionary by assigning a value to a new key.

->Using Assignment:

#Create an empty dictionary

person = {}

#Add new key-value pairs

person["name"] = "Alice"

person["age"] = 30

person["city"] = "New York"

print(person)

#Output: {'name': 'Alice', 'age': 30, 'city': 'New York'}

->Using the update() Method:

You can also add multiple key-value pairs at once using the update() method. person = {"name": "Alice"}

#Add multiple items

person.update({"age": 30, "city": "New York"})

print(person)

#Output: {'name': 'Alice', 'age': 30, 'city': 'New York'}

ii. Modifying Items

You can modify the value associated with an existing key by reassigning a new value to that key.

->Using Assignment

person = {"name": "Alice", "age": 30, "city": "New York"}

#Modify the value of an existing key

person["age"] = 31

person["city"] = "San Francisco"

print(person)

#Output: {'name': 'Alice', 'age': 31, 'city': 'San Francisco'}

iii. Deleting Items

You can remove key-value pairs from a dictionary using several methods.

->Using the del Statement:

The del statement removes a key-value pair based on the key.

person = {"name": "Alice", "age": 31, "city": "San Francisco"}

#Remove an item by key

del person["city"]

print(person)

#Output: {'name': 'Alice', 'age': 31}

->Using the pop() Method:

The pop() method removes a key-value pair and returns the value associated with the key. If the key is not found, it raises a KeyError

unless a default value is provided.

person = {"name": "Alice", "age": 31, "city": "San Francisco"}

#Remove an item by key and return its value

city = person.pop("city")

print(city) # Output: San Francisco

print(person) # Output: {'name': 'Alice', 'age': 31}

#Remove an item with a default value if the key is not found

country = person.pop("country", "Not Found")

print(country) # Output: Not Found

->Using the clear() Method:

The clear() method removes all key-value pairs from the dictionary, leaving it empty.

person = {"name": "Alice", "age": 31, "city": "San Francisco"}

#Remove all items

person.clear()

print(person) # Output: {}

8 Discuss the importance of dictionary keys being immutable and provide examples

Ans

In Python, the keys of a dictionary must be immutable. This is because dictionary keys are used to generate hash values, which are then used to quickly retrieve the corresponding values. If a key were mutable, its hash value could change after it was added to the dictionary, which would make it impossible to find the key later, leading to inconsistencies and errors.

Importance of Immutable Keys

Hashability: The primary reason dictionary keys need to be immutable is that they must be hashable. The hash function is used to determine the index at which the value associated with a particular key is stored. Immutable objects have a fixed hash value, which ensures that the key can be reliably found in the dictionary.

Consistency: If keys were mutable, changing a key after it was added to a dictionary would lead to the dictionary becoming inconsistent. For instance, the key could no longer be located, and the integrity of the data structure would be compromised.

Performance: The immutability of keys contributes to the efficiency of dictionary operations like lookups, insertions, and deletions. Since the hash value of a key remains constant, these operations can be performed in average O(1) time.

Examples

Immutable Keys:

-> Strings:

Strings are immutable in Python, so they are commonly used as dictionary keys

my_dict = {"name": "Alice", "age": 30}

-> Tuples:

Tuples can also be used as dictionary keys, provided all the elements within the tuple are themselves immutable.

coordinates = {(40.7128, -74.0060): "New York", (34.0522, -118.2437): "Los Angeles"}

Mutable Keys (Not Allowed):

-> Lists:

Lists are mutable and therefore cannot be used as dictionary keys. Attempting to do so will raise a TypeError.

my_dict = {[1, 2, 3]: "List"} # This will raise an error

-> Dictionaries:

Dictionaries themselves are mutable and cannot be used as keys for another dictionary.

my_dict = {{"key": "value"}: "Dictionary"} # This will raise an error
